# Week 2 - Image Preprocessing and Tesseract OCR

This notebook preprocesses Urdu OCR images by converting them to grayscale, resizing them, removing noise, and binarizing them. It also evaluates Tesseract OCR on Urdu text images to understand its limitations.

```python
!pip install opencv-python-headless pillow

import cv2
import numpy as np
from PIL import Image
import os
import matplotlib.pyplot as plt

print("Libraries loaded successfully!")
```

```python
def preprocess_image(image_path, save_path):
    # Load image
    img = cv2.imread(image_path)

    if img is None:
        print(f"Could not load: {image_path}")
        return

    # Step 1: Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Step 2: Resize image
    resized = cv2.resize(gray, (512, 128))

    # Step 3: Remove noise
    denoised = cv2.fastNlMeansDenoising(resized, h=10)

    # Step 4: Convert to black & white
    _, binary = cv2.threshold(denoised, 127, 255, cv2.THRESH_BINARY)

    # Save processed image
    cv2.imwrite(save_path, binary)

    return binary


# Create processed folder
os.makedirs("urdu-ocr-codesaviours-si26-saima/data/processed", exist_ok=True)

print("Preprocessing function ready!")
```

```python
import glob
import os

# Find all images
all_images = glob.glob(
    "urdu-ocr-codesaviours-si26-saima/data/raw/**/*.jpg",
    recursive=True
)

all_images += glob.glob(
    "urdu-ocr-codesaviours-si26-saima/data/raw/**/*.png",
    recursive=True
)

print(f"Found {len(all_images)} images to process")

processed_count = 0

for img_path in all_images:

    filename = os.path.basename(img_path)

    save_path = f"urdu-ocr-codesaviours-si26-saima/data/processed/{filename}"

    result = preprocess_image(img_path, save_path)

    if result is not None:
        processed_count += 1

print(f"Done! Processed {processed_count} images")
print("Check data/processed/ folder")
```
