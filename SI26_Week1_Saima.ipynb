# What is OCR?

Optical Character Recognition (OCR) is a technology that converts text from images, scanned documents, or photographs into editable and machine-readable text. It helps computers recognize printed or handwritten characters automatically. OCR is widely used for digitizing books, forms, and historical documents.

# Why is Urdu OCR harder than English OCR?

Urdu OCR is more difficult because Urdu is written in a cursive script where letters are connected. The shape of a letter changes depending on its position in a word. Different fonts, ligatures, and image quality also make Urdu text recognition more challenging than English.
