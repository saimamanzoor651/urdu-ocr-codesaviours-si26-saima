# What is OCR?

Optical Character Recognition (OCR) is a technology that converts text from images, scanned documents, or photographs into editable and machine-readable text. It helps computers recognize printed or handwritten characters automatically. OCR is widely used for digitizing books, forms, and historical documents.

# Why is Urdu OCR harder than English OCR?

Urdu OCR is more difficult because Urdu is written in a cursive script where letters are connected. The shape of a letter changes depending on its position in a word. Different fonts, ligatures, and image quality also make Urdu text recognition more challenging than English.

1. Converting old Urdu books and newspapers into searchable digital text for preservation.

2. Extracting text from Urdu signboards, forms, and documents for digital record keeping and translation.

```python
!pip install Pillow arabic-reshaper python-bidi

from PIL import Image, ImageDraw, ImageFont
import arabic_reshaper
from bidi.algorithm import get_display
import os

urdu_texts = [
    "پاکستان زندہ باد",
    "آج کا موسم خوشگوار ہے",
    "تعلیم ہر انسان کا حق ہے",
    "کراچی پاکستان کا سب سے بڑا شہر ہے",
    "محنت کامیابی کی کنجی ہے"
]

os.makedirs("data/synthetic", exist_ok=True)

for i, text in enumerate(urdu_texts):
    reshaped = arabic_reshaper.reshape(text)
    bidi_text = get_display(reshaped)

    img = Image.new("RGB", (600, 120), color="white")
    draw = ImageDraw.Draw(img)

    draw.text((20, 40), bidi_text, fill="black")

    img.save(f"data/synthetic/urdu_{i+1}.png")

    print(f"Generated urdu_{i+1}.png")

print("Done! Check data/synthetic/")
```
